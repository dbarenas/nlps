```python
#remove punctuation and numbers from a string
#returns : clean list
from nltk.tokenize import RegexpTokenizer

def clean(text):
    tst=text.lower()
    tokenizer = RegexpTokenizer(r'\w+')
    r=tokenizer.tokenize(tst)
    res=[]
    for i in r:
        if i.isdigit() is not True:
            res.append(i)
     
    return " ".join(res)
```
